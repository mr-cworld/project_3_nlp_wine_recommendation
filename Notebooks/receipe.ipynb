## Roles

```python
# Additional imports for human and system messages.
from langchain.schema import HumanMessage, SystemMessage

# Import ChatGoogleGenerativeAI
from langchain_google_genai import ChatGoogleGenerativeAI 
```

```python
# Initialize the model.
#llm = ChatGoogleGenerativeAI(google_api_key=GEMINI_API_KEY, model=GEMINI_MODEL, temperature=0.3)
llm = ChatGoogleGenerativeAI(google_api_key=GEMINI_API_KEY, model=GEMINI_MODEL, temperature=0.3)


# Format the messagae content
message_content = f"Provide me with a suggested meal that pairs well with a {color} wine that has {categories} categories and {aromas_aroma} aromas. Include the recipe to cook the meal and the shopping list."

# Create the HumanMessage
human_message = HumanMessage(content=message_content)

# Create a list containing a system message and a human message.
messages = [
    SystemMessage(content="You are a culinary chef."),
    human_message
]

# Provide the messages to the LLM and print the result. Replace asterisk with nothing on the print results
#print(llm.invoke(messages).content.replace("*", ""))
result = llm.invoke(messages)
result=result.content.replace("*", "")
print(result)
```

## Parse Introduction, Receipe & Ingredients, and Shopping List to pass to Gradio App

```python
import re

# Extract the full content from the AI response
content = result

# Extract the introduction (text before "Recipe:")
intro_match = re.search(r"^(.*?)Recipe:", content, re.DOTALL)

# Extract the recipe and ingredients (text between "Recipe:" and "Shopping List:")
recipe_match = re.search(r"Recipe:\s*(.*?)Shopping List:", content, re.DOTALL)

# Extract the shopping list and separate extra content
shopping_list_match = re.search(r"Shopping List:\s*\n([\s\S]+)", content)

shopping_list_text = shopping_list_match.group(1).strip() if shopping_list_match else ""

# Identify the actual shopping list by detecting where item formatting changes
shopping_list_lines = shopping_list_text.split("\n")

# Find the index where the shopping list stops and extra content begins
shopping_list = []
extra_content = []
found_extra = False

for line in shopping_list_lines:
    if found_extra or not re.match(r"^\s*[\w\d(-]", line):  # Detects non-item lines
        found_extra = True
        extra_content.append(line)
    else:
        shopping_list.append(line.strip())

# Convert extra content back into a single string
extra_text = "\n".join(extra_content).strip()

# Store the extracted introduction
intro_text = intro_match.group(1).strip() if intro_match else ""

# Store the extracted recipe & ingredients
recipe_text = recipe_match.group(1).strip() if recipe_match else ""

# Print the results
print("INTRODUCTION:")
print(intro_text)
print("\nRECIPE & INGREDIENTS:")
print(recipe_text)
print("\nShopping List:\n")
for item in shopping_list:
    print(f"* {item}")
print("\n")
print(extra_text)
```
