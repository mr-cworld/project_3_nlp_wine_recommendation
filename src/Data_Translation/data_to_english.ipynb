```python
#Libraries 
import pandas as pd
```

```python
#Using above df.info(), we can see most columns have null values.
#Drop null values for these columns, ignoring Wine_bottle_label as the Review_Notes are most important (specific to each bottle by a judge), and 1026 rows are about 20% of dataset
df.dropna(subset=['Name','Year','Color','Castes','Winery','Min_Price','Max_Price','Judge','Judge_Rating','Review_Date','Review_Notes'], inplace=True)
```

```python
#Checking it worked
df.info()

#it worked, about 300 rows were dropped
```

```python
#New Library
%pip install googletrans
from googletrans import Translator
```

```python
import asyncio

#Create Translator object
translator = Translator()
#Declare my columns to translate
columns_to_translate = ['Review_Notes', 'Wine_Bottle_Label']

#For loop to translate all the columns, if no value leave as is (for the 1k null wine labels)
async def translate_text(text):
    if pd.notna(text):
        translated = await translator.translate(text, src='pt', dest='en')
        return translated.text
    return text

async def translate_columns(df, columns):
    for col in columns:
        if col in df.columns:
            df[col] = await asyncio.gather(*[translate_text(x) for x in df[col]])
    return df

df = await translate_columns(df, columns_to_translate)

"""
for col in columns_to_translate:
    if col in df.columns:
        df[col] = df[col].apply(lambda x: translator.translate(x, src='pt', dest='en').text if pd.notna(x) else x)"""
```
